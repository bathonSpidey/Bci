```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
left_data=pd.read_csv('left.csv')
```

```python
left_data.head()
```

```python
left_data=left_data.drop(['Time'], axis=1)
```

```python
left_data.head()
```

```python
left_data['Label']=0
```

```python
left_data.head()
```

```python
right_data=pd.read_csv('right.csv')
```

```python
right_data.head()
```

```python
right_data=right_data.drop(['Time'], axis=1)
```

```python
right_data['Label']=1
```

```python
right_data.head()
```

```python
merged_data = pd.concat([left_data, right_data])
```

```python
merged_data.head(-5)
```

```python
y=merged_data['Label']
```

```python
y
```

```python
cols=merged_data.columns
print(type(cols))
for i in cols:
    sns.catplot(x="Label", y=i, data=merged_data)
```

```python
train_data=merged_data.drop(['Label'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data,y,test_size=0.30, random_state=121)
```

```python
from sklearn.svm import SVC
model_svc=SVC(gamma='scale',degree=4)
model_svc.fit(X_train,y_train)
pred_svc=model_svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred_svc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_svc))
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test, pred_svc, pos_label=2)
```

```python
fpr
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, pred_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier(random_state=121)
model_dtc.fit(X_train,y_train)
pred_dtc=model_dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_dtc))
```

```python
roc_auc_score(y_test, pred_dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rfc = RandomForestClassifier(n_estimators=500, max_depth=2,random_state=40)
model_rfc.fit(X_train,y_train)
pred_rfc=model_rfc.predict(X_test)
```

```python
print(classification_report(y_test,pred_rfc))
```
