```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
left_data=pd.read_csv('left.csv')
left_data.head()
```

```python
left_data=left_data.drop(['Time'], axis=1)
left_data.head()
```

```python
left_data['Label']=0
left_data.head()
```

```python
right_data=pd.read_csv('right.csv')
right_data.head()
```

```python
right_data=right_data.drop(['Time'], axis=1)
right_data['Label']=1
right_data.head()
```

```python
up_data=pd.read_csv('up.csv')
up_data.head()
```

```python
up_data=up_data.drop(['Time'], axis=1)
up_data['Label']=2
up_data.head()
```

```python
down_data=pd.read_csv('down.csv')
down_data.head()
```

```python
down_data=down_data.drop(['Time'], axis=1)
down_data['Label']=3
down_data.head()
```

```python
merged_data = pd.concat([left_data, right_data, up_data, down_data])
merged_data.head(-1)
```

```python
y=merged_data['Label']
y
```

```python
cols=merged_data.columns
print(type(cols))
for i in cols:
    sns.catplot(x="Label", y=i, data=merged_data)
```

```python
train_data=merged_data.drop(['Label'], axis=1)
train_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data,y,test_size=0.30, random_state=121)
```

```python
from sklearn.svm import LinearSVC
model_svc=LinearSVC()
model_svc.fit(X_train,y_train)
pred_svc=model_svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_svc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=10,algorithm='brute')
neigh.fit(X_train, y_train) 
pred=neigh.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
