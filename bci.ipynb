```python
y=data_frame['Frequency']
X=data_frame
```

```python
X
```

Scaling the data

```python
X.drop("Frequency", axis = 1, inplace = True)
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
```

```python
transform_X=scaler.transform(X)
```

```python
transform_X
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(transform_X, y, test_size=0.33, random_state=42)
```

Some classification algorithms:
1. SVC
2. Decision Tree
3. Random Forest
4. Linear Classification

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train, y_train)
pred=svc.predict(X_test)
```
